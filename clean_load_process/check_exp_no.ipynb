```python
import sqlite3 
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
conn = sqlite3.Connection('../data/processed/fire_data.db')
```

```python
df = pd.read_sql("""
    SELECT *
    FROM basic_incident
    LIMIT 1000000
""", conn)
```

```python
df[df.ALARM == 62220200143]
```

```python
display(df \
    .loc[:, ['STATE', 'FDID', 'INC_DATE', 'INC_NO', 'EXP_NO']] \
    .groupby(by=['STATE', 'FDID', 'INC_DATE', 'INC_NO']) \
    .max() \
    .reset_index() \
    .EXP_NO \
    .mean())

display(df \
    .loc[:, ['STATE', 'FDID', 'INC_DATE', 'INC_NO', 'EXP_NO']] \
    .groupby(by=['STATE', 'FDID', 'INC_DATE', 'INC_NO']) \
    .size() \
    .reset_index() \
    .iloc[:, -1] \
    .mean())
```
